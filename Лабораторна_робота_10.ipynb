```python
import os
import random

filename = 'mas.txt'

if not os.path.exists(filename):

    random_numbers = [str(random.randint(-100, 100)) for _ in range(10)]
    with open(filename, 'w') as file:
        file.write(' '.join(random_numbers))
    print(f"Файл '{filename}' був створений з такими числами:")
    print(' '.join(random_numbers))

with open(filename, 'r') as file:
    data = file.read()

numbers = list(map(int, data.split()))

positive_numbers = [num for num in numbers if num > 0]

if positive_numbers:
    average = sum(positive_numbers) / len(positive_numbers)
    print(f"Середнє арифметичне додатних чисел: {average}")
else:
    print("У файлі немає додатних чисел для обчислення середнього арифметичного.")
```
